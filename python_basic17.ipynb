```python
# Question 3
def replace_vowels(a,b):
    c=['a','e','i','o','u']
    d=''
    for i in a:
        if i in c:
            i=b
            d=d+i
        else:
            d=d+i
    return d
a = input("enter string : ")
b = input("enter character: ")
r = replace_vowels(a, b) 
print(r)
```

```python
# Question 4

def factorial(a):
    if a==1:
        return a
    else:
        return a * factorial(a-1)  

a = int(input("enter a number : ") )
b = factorial(a)
print(b)
```

```python
# Question 5
def hamming_distance(a,b):
    c=0
    if len(a)== len(b):
        for i in range(len(a)):
             if a[i]!=b[i]:
                c=c+1
        return c
    else:
        return "length of two string is not same try again"
a = input("enter 1st string : ") 
b = input("enter 2nd string : ")
c = hamming_distance(a,b)
if type(c)==int:
    print(c)
else:
    print(c)
```
